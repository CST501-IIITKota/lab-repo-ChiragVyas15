```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons
```

```python
data=make_moons(n_samples=300,noise=0.05,random_state=0)[0]
```

```python
data
```

```python
arr=np.array(data)
```

```python
arr
```

```python
#[plot this data
plt.scatter(arr[:,0],arr[:,1])
plt.show()
```

```python
dbscan=DBSCAN(eps=0.2,min_samples=5)
dbscan.fit(arr)
labels=dbscan.labels_
print(labels)
```

```python
#plot the clustered data
colors=['r.','g.','b.','c.','k.','y.']
for i in range(len(arr)):
    plt.plot(arr[i][0],arr[i][1],colors[labels[i]],markersize=10)
plt.show()
```

```python
#define core ,border and noise points
core_samples_mask = np.zeros_like(dbscan.labels_, dtype=bool)
core_samples_mask[dbscan.core_sample_indices_] = True
core_points = arr[core_samples_mask]
border_points = arr[(dbscan.labels_ != -1) & ~core_samples_mask]
noise_points = arr[dbscan.labels_ == -1]
```

```python
#plot core, border and noise points
plt.scatter(core_points[:, 0], core_points[:, 1], c='black', label='Core Points')
plt.scatter(border_points[:, 0], border_points[:, 1], c='green', label='Border Points')
plt.scatter(noise_points[:, 0], noise_points[:, 1], c='red', label='Noise Points')
plt.legend()
plt.show()
```
